# K Fold CV(Cross Validation) 모델 성능평가
최종 성능은 각 실험 성능의 평균으로 도출

```python
# (중요) 컬럼이 feature, 로우가 sample은 국룰

from sklearn.model_selection import KFold
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import numpy as np

iris = load_iris()
features = iris.data
label = iris.target

dt_clf = DecisionTreeClassifier(random_state = 156) # random_state가 완전랜덤이 아닌 패턴이 있는 랜덤으로 돌리게 해줌 

# n_splits개의 폴드 세트로 분리하는 KFold 객체와 폴드 세트별 정확도를 담을 리스트 객체 선언
kfold = KFold(n_splits=5)
cv_accuracy = []                                     # cross validation을 하면 여러 개의 값이 나오므로 리스트로 만드는 것이 중요
print("붓꽃 데이터 세트 크기 : ", features.shape[0]) # features.shape[i]는 샘플의 개수임을 '기억'
print("=" * 80)

n_iter = 0 # 회수 출력을 위한 변수
# kfold.split()으로 반환된 인덱스를 이용하여 학습용, 검증용 테스트 데이터 추출
for train_index, test_index in kfold.split(features) : 
    X_train, X_test = features[train_index], features[test_index]
    y_train, y_test = label[train_index], label[test_index]

    dt_clf.fit(X_train, y_train)
    pred = dt_clf.predict(X_test)
    print("예측값 : %s " % pred)
    n_iter += 1
    accuracy = np.round(accuracy_score(y_test, pred), 4)
    
    # 트레이닝 피쳐와 테스트 피쳐의 행 개수를 의미
    train_size = X_train.shape[0]
    test_size = X_test.shape[0]
    print("\n#{0} 교차 검증 정확도 : {1}, 학습 데이터 크기 : {2}, 검증 데이터 크기 : {3}"
         .format(n_iter, accuracy, train_size, test_size))
    print("#{0} 검증 세트 인덱스 : {1}".format(n_iter, test_index))
    cv_accuracy.append(accuracy)
    
print("\n# 평균 검증 정확도 : ", np.mean(cv_accuracy)) # 개별 정확도를 합하여 평균 정확도 계산
```

# Stratified K Fold CV(Cross Validation)
Stratified + K Fold CV. 즉, K Fold CV를 하는데 

```python
from sklearn.model_selection import StratifiedKFold
import pandas as pd

iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['label'] = iris.target

print("< iris_df['label'].value_counts() >\n", iris_df['label'].value_counts()) # .value_count() 는 컬럼의 값별 개수를 출력
print("\n< iris_df['label'] >\n", iris_df['label'])

skf = StratifiedKFold(n_splits = 3) # 데이터 셋을 3개로 나눔
n_iter = 0

# 이미 n_splits개로 나뉘어진 Stratified K Fold 객체에 데이터와 나눌 값이 저장된 컬럼을 넣는다
for train_index, test_index in skf.split(iris_df, iris_df['label']) :
    n_iter += 1
    label_train = iris_df['label'].iloc[train_index]
    label_test = iris_df['label'].iloc[test_index]
    print("\n## 교차 검증 : {0}" .format(n_iter))
    print("학습 레이블 데이터 분포 : \n", label_train.value_counts())
    print("\n검증 레이블 데이터 분포 : \n", label_test.value_counts())

df_clf = DecisionTreeClassifier(random_state = 156)
skfold = StratifiedKFold(n_splits=3)
n_iter = 0
cv_accuracy = []

# StratifiedFold split() 호출 시 반드시 레이블 데이터 셋도 추가 입력 필요
for train_index, test_index in skfold.split(features, label) :
    X_train, X_test = features[train_index], features[test_index]
    y_train, y_test = label[train_index], label[test_index]
    dt_clf.fit(X_train, y_train)
    pred = dt_clf.predict(X_test)
    
    n_iter += 1
    accuracy = np.round(accuracy_score(y_test, pred), 4)
    train_size = X_train.shape[0]
    test_size = X_test.shape[0]
    print("\n#{0} 교차 검증 정확도 : {1}, 학습 데이터 크기 : {2}, 검증 데이터 크기 : {3}"
         .format(n_iter, accuracy, train_size, test_size))
    print("#{0} 검증 세트 인덱스 : {1}".format(n_iter, test_index))
    
    cv_accuracy.append(accuracy)

# 교차 검증별 정확도 및 평균 정확도 계산
print('\n# 교차 검증별 정확도:', np.round(cv_accuracy, 4))
print("# K Fold보다 정확도 향상됨.")
```

# 위의 Stratified K Fold를 cross_val_score 모듈을 통해 보다 간단하게 구현
* cross_val_score() : cross validation 결과의 정확도를 측정

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score, cross_validate
from sklearn.datasets import load_iris

iris_data = load_iris()
dt_clf = DecisionTreeClassifier(random_state = 156)

data = iris_data.data
label = iris_data.target

# 성능 지표는 정확도(accuracy), 교차 검증 세트는 3개
scores = cross_val_score(dt_clf, data, label, scoring = 'accuracy', cv = 3)
print("교차 검증별 정확도 : ", np.round(scores, 4))
print("평균 검증 정확도 : ", np.round(np.mean(scores), 4))
```

# GridSearch CV : 파라미터를 바꿈으로써 최적의 파라미터 조합을 쉽게 찾을 수 있음
* max_depth : 트리의 최대 깊이 지정. 지정하지 않으면 깊이가 제한되지 않아 overfitting이 발생할 확률이 높으므로 중요한 파라미터임
* min_samples_split  : 규제를 걸어 overfitting이 되지 않게 해줌
* refit = True 가 default임. True이면 가장 좋은 파라미터 설정으로 재학습 시킴

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
import pandas as pd


iris_data = load_iris()

X_train, X_test, y_train, y_test = train_test_split(iris_data.data, iris_data.target, test_size = 0.2, random_state = 121)
dtree = DecisionTreeClassifier()
parameters = {'max_depth' : [1,2,3], 'min_samples_split' : [2,3]}

# param_grid의 하이퍼 파라미터들을 3개의 train, test set fold로 나누어서 테스트 수행 설정.
grid_dtree = GridSearchCV(dtree, param_grid=parameters, cv=3, refit=True)

grid_dtree.fit(X_train, y_train) # Train 데이터로 param_grid의 하이퍼 파라미터들을 순차적으로 학습/평가

# GridSearchCV 결과 추출하여 DataFrame으로 변환
# cv_results_ 변수에 grid_dtree.fit(X_train, y_train) 의 결과값이 들어 있음
scores_df = pd.DataFrame(grid_dtree.cv_results_)

display(scores_df[['params', 'mean_test_score', 'rank_test_score', 'split0_test_score', 'split1_test_score', 'split2_test_score']])
print('GridSearchCV 최적 파라미터 : ', grid_dtree.best_params_)
print('GridSearchCV 최고 정확도 : {0:.4f}' .format(grid_dtree.best_score_))

# GridSearchCV의 refit으로 이미 학습이 된 estimator 변환
estimator = grid_dtree.best_estimator_

# GridSearchCV의 best_estimator_는 이미 최적 하이퍼 파라미터로 학습이 됨
pred = estimator.predict(X_test)
print('테스트 데이터 정확도 : {0:.4f}' .format(accuracy_score(y_test, pred)))
```
