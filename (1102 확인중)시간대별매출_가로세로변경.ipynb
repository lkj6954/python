### 날짜를 각 시간대의 열으로 넣기

```python
df4 = df3.copy()
col4 = df4.columns.tolist()

for num in range(0,len(col4)) : 
    df4['날짜'+str(num)] = col4[num]
    col_n = df4.columns.tolist()
    tmp = col_n[len(col_n)-1]
    del col_n[len(col_n)-1]
    df4.drop(['날짜'+str(num)],axis='columns')
#     df4.drop([df4.index[len(col4)]])
    col_n.insert(2*num,tmp)
    df4=df4[col_n]
df4.head()
```

### 시간(0~23)순으로 쭉 나열 : column 두 개씩 뽑아서 아래에 붙이기 

```python
refine_df = pd.DataFrame()
for i in range(1, len(df4.columns), 2):
    temp_df = pd.DataFrame(df4.iloc[:, i])
    temp_df.columns = ['매출액']
    refine_df = pd.concat([refine_df, temp_df])
    print(i, '/', len(df4.columns))
```

```python
date_df = pd.DataFrame()
for i in range(0, len(df4.columns), 2):
    temp_df = pd.DataFrame(df4.iloc[:, i])
    temp_df.columns = ['날짜']
    date_df = pd.concat([date_df, temp_df])
```

```python
final = pd.concat([date_df, refine_df], axis = 1)
```
