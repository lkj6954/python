```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import rhinoMorph
rn = rhinoMorph.startRhino()

# 그래프에서 한글표현을 위해 폰트를 설치
%config InlineBackend.figure_format = 'retina'       # 폰트가 깨끗하게 보이도록 설정

import matplotlib.font_manager as fm
fontpath = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
font = fm.FontProperties(fname=fontpath, size=9)

# 기본 글꼴 변경
mpl.pyplot.rc('font', family='NanumGothic')
mpl.font_manager._rebuild()
```

```python
# 데이터 로딩
def read_data(filename, encoding='cp949'):                      # 읽기 함수 정의
    with open(filename, 'r', encoding = encoding) as f:   
        return f.read()
```

```python
# 텍스트파일 불러오기
dataKim = read_data('C:/Users/LeeKyoungJun/Desktop/데이터캠퍼스/pytest/김소월시.txt', encoding='cp949')
dataYoon = read_data('C:/Users/LeeKyoungJun/Desktop/데이터캠퍼스/pytest/윤동주시.txt', encoding='cp949')

# 형태소만 남기기
kimMorph = rhinoMorph.onlyMorph_list(rn, dataKim, pos=['NNG', 'NNP', 'VV', 'VA', 'XR', 'IC', 'MM', 'MAG', 'MAJ'], eomi=True)
yoonMorph = rhinoMorph.onlyMorph_list(rn, dataYoon, pos=['NNG', 'NNP', 'VV', 'VA', 'XR', 'IC', 'MM', 'MAG', 'MAJ'], eomi=True)
 
# 카운트 써서 값 계산
from collections import Counter
kimCount = Counter(kimMorph) # 딕셔너리 형태
yoonCount = Counter(yoonMorph) # 딕셔너리 형태
```

```python
import matplotlib.pyplot as plt

kimSortedKeys = sorted(kimCount, key=kimCount.get, reverse=True)
yoonSortedKeys = sorted(yoonCount, key=yoonCount.get, reverse=True)
kimSortedValues = sorted(kimCount.values(), reverse=True)
yoonSortedValues = sorted(yoonCount.values(), reverse=True)

kimRatioValues = []
for i in kimSortedValues :
    kimRatioValues.append(round(i / sum(kimSortedValues), 2))
kimRatioValues[:20]

yoonRatioValues = []
for i in yoonSortedValues :
    yoonRatioValues.append(round(i / sum(yoonSortedValues), 2))
yoonRatioValues[:20]

print("김소월 단어빈도 총합 : %s" % len(kimMorph) )
print("김소월 고빈도 단어 : %s" % kimSortedKeys[:20])
print("김소월 고빈도 단어비율 : %s\n" % kimRatioValues[:20])
print("=== 김소월 단어비율 ===")
plt.bar(range(20), kimSortedValues[:20])
plt.xticks(range(20), kimSortedKeys[:20]) # 그래프 하단에 문자열 출력
plt.show()

print("윤동주 단어빈도 총합 : %s" % len(yoonMorph) )
print("윤동주 고빈도 단어 : %s" % yoonSortedKeys[:20])
print("윤동주 고빈도 단어비율 : %s\n" % yoonRatioValues[:20])
print("=== 윤동주 단어비율 ===")
plt.bar(range(20), yoonSortedValues[:20])
plt.xticks(range(20), yoonSortedKeys[:20]) # 그래프 하단에 문자열 출력
plt.show()
```
