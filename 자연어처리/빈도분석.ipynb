```python
import matplotlib as mpl
import matplotlib.pyplot as plt

# 그래프에서 한글표현을 위해 폰트를 설치
%config InlineBackend.figure_format = 'retina'       # 폰트가 깨끗하게 보이도록 설정

import matplotlib.font_manager as fm
fontpath = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
font = fm.FontProperties(fname=fontpath, size=9)

# 기본 글꼴 변경
mpl.pyplot.rc('font', family='NanumGothic')
mpl.font_manager._rebuild()
```

```python
# 데이터 로딩
def read_data(filename, encoding='cp949'):                      # 읽기 함수 정의
  with open(filename, 'r', encoding = encoding) as f:
    data = [line.split('\t') for line in f.read().splitlines()] # line comprehensioon
    data = data[1:]                                             # txt 파일의 헤더(id document label)는 제외하기
  return data

def write_data(data, filename, encoding='cp949'):               # 쓰기 함수도 정의
  with open(filename, 'w', encoding=encoding) as f:
    f.write(data)

# data = read_data('/content/gdrive/My Drive/pytest/ratings_small.txt', encoding='cp949')
data = read_data('C:/Users/LeeKyoungJun/Desktop/데이터캠퍼스/pytest/ratings_small.txt', encoding='cp949')  # 전체파일은 ratings.txt (긍정 1만, 부정 1만)
```

```python
# 데이터 확인
print("===== data 변수 전체 출력 =====")
print(data)

print("data 변수 길이 : %s" % len(data))
print("data의 0번째 리스트의 원소 개수 : ", len(data[0]))
print("data의 0번째 리스트의 원소 : ", data[0])
print("data의 0번째에서 2번째 리스트의 원소 : ", data[0:3])
```

```python
# 샘플 데이터 분석 연습
import rhinoMorph
rn = rhinoMorph.startRhino()

# 형태소 분석된 문장 샘플 보기
# 두번째 인자는 스트링이 들어가야 함
sample_data = rhinoMorph.onlyMorph_list(rn, data[0][1], pos=['NNG', 'NNP', 'VV', 'VA', 'XR', 'IC', 'MM', 'MAG', 'MAJ'], eomi=True)

print("원본 : %s" % data[0][1])
print('sample data:', sample_data)          # 형태소 분석 결과
print('joined sample data:', ' '.join(sample_data))	  # 문자열을 공백으로 연결한다
```

```python
# 전체 데이터 형태소 분석

morphed_data = ''
for data_each in data: 
  # data_each[1] 부분이 어떻게 for문에서 작동되는지 고민했었음.
  # list = [1,2,3,4,5]
  # for i in list :
  #   print(i)
  # 이게 작동되는 것과 동일하다고 이해하면 됨. data가 이중 리스트라서 어색했던 거임

  morphed_data_each = rhinoMorph.onlyMorph_list(rn, data_each[1], pos=['NNG', 'NNP', 'VV', 'VA', 'XR', 'IC', 'MM', 'MAG', 'MAJ'], eomi=True)
  joined_data_each = ' '.join(morphed_data_each)        # 문자열을 하나로 연결
  if joined_data_each :                                  # 내용이 있는 경우만 저장하게 함
    morphed_data += data_each[0]+"\t"+joined_data_each+"\t"+data_each[2]+"\n"
    
# 형태소 분석된 파일 저장
write_data(morphed_data, 'ratings_morphed.txt', encoding='cp949')
print(morphed_data)
```

```python
# 분석된 데이터 로딩

# 위에서 morphed_data는 str형태인데, 받아오니까 list 형태가 됨
data = read_data('ratings_morphed.txt', encoding='cp949')
print("data의 리스트 개수 : ", len(data))                              # 일부는 내용이 남지 않아 제외 됨
print("data[0] : %s" % data[0])
print("data[0]의 길이 : %s" % len(data[0]))

data_text = [line[1] for line in data]      	# 데이터 본문
data_senti = [line[2] for line in data]     	# 데이터 긍부정 부분 (1은 긍정, 0은 부정)
```

```python
# counter 연습
from collections import Counter
count = Counter(["여름", "과일", "딸기", "과일", "딸기"]) # 리스트에 포함된 단어의 수만큼 카운팅
print("카운팅 결과 : %s" % count)
result = count.most_common(2) # 가장 출현 빈도수가 높은 n개를 저장

print(result)
print(result[0])
print(result[0][0])		# 0 번째 요소의 0 번째
print(result[0][1])		# 0 번째 요소의 1 번째
```

```python
# 분석된 데이터 빈도 구하기

print('< 본문 리스트(data_text) >')            # data_text : 리스트 data에 추가한 데이터 본문 컬럼
print(data_text)
mergedText = ' '.join(data_text)         	# 좋은 방법. 공백을 추가하며 모든 리스트 요소들을 결합한다
print('< 리스트들을 합쳐 문자열로 만든 것(mergedText) >')
print(mergedText)

# 리스트로 되어있던 걸 풀어서 다시 합치고 공백 기준으로 자른 이유는 기존의 리스트(data_text)는 형태소 단위로 잘려진 문장이었는데, 이 형태에서는 단어 카운팅을 할 수 없다.
# 따라서 공백 단위로 나누어야 되는데, 어차피 모든 리스트의 값을 공백으로 나눠야되니 모두 다 합친 뒤에 공백 단위로 자른 것이다.

mergedTextList = mergedText.split(' ')   	# 결합된 요소들을 공백 단위로 분리하여 하나의 리스트로 만든다
print('mergedTextList:', mergedTextList)

wordInfo = Counter(mergedTextList)  	    # 하나의 리스트로 결합된 요소를 카운트한다 (내림차순)
print('wordInfo:', wordInfo)
```

```python
# sorted 연습

sample = {'여름':1, '과일':2, '딸기':3}
print("sorted(sample) : %s " % sorted(sample))
print("sorted(sample, reverse=True) : %s " % sorted(sample, reverse=True))                     # 역순으로 정렬한다
print("sorted(sample, key=sample.get, reverse=True) : %s " % sorted(sample, key=sample.get, reverse=True))     # sample.get의 출력된 값을 기준으로 sample을 정렬한다. key=sample.get 으로 빈도별 출력이 됨.
print("sorted(sample.values(), reverse=True) : %s" % sorted(sample.values(), reverse=True))
```

```python
# bar 그래프 연습

wordInfo_sample = Counter({'여름':1, '과일':2, '딸기':3})
sorted_keys_sample = sorted(wordInfo_sample, key=wordInfo_sample.get, reverse=True)
sorted_values_sample = sorted(wordInfo_sample.values(), reverse=True)

import matplotlib.pyplot as plt
plt.bar(range(len(wordInfo_sample)), sorted_values_sample) # X축의 위치, 각 x의 높이
plt.xticks(range(len(wordInfo_sample)), sorted_keys_sample) # X축의 위치, 각 x의 라벨
plt.show()
```
