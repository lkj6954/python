# Series 연습

```python
import pandas as pd
temp = pd.Series(['홍길동', '전우치', '강감찬', '스티브잡스'])
temp
```

```python
# 위 결과에서 인덱스 변경

# 이렇게도 됨
# temp.index = ['1번', '2번', '3번', '4번']
# temp

temp = pd.Series(['홍길동', '전우치', '강감찬', '스티브잡스'], index = ['1번', '2번', '3번', '4번'])
temp
```

## 이미 만들어진 Series에서 temp.index = ['1번', '2번', '3번', '4번'] 이렇게로도 인덱스 변경 가능

```python
# 딕셔너리 형으로 Series 생성하기

tempDic = {'일지매' : 120, '전우치' : 100, '홍길동' : 80}
temp = pd.Series(tempDic)
temp
```

## Series를 딕셔너리 형으로 생성할 때는 딕셔너리로 선언을 하고, 그 값을 Series 선언할 때 인자로 넣어야 한다. 그리고 Key값이 Series의 인덱스가 된다.

```python
# Series 전체 조회
print(temp)
print("\n")

# 행 번호로 조회
print(temp[0])
print("\n")

# 인덱스 명으로 조회
print('홍길동')
print("\n")

# 행 번호와 인덱스 명으로 여러 로우 조회
print(temp[ [0,1] ])
print("\n")
print(temp[ ['전우치', '홍길동'] ])
```

## Series형 값 조회할 때는 인덱스값 또는 행번호로.

```python
# Series끼리 연산

temp2 = temp
print(temp + temp2)
print("\n")

# 서로 인덱스가 겹치지 않을 때 연산
tempDic = {'이경준' : 100, '홍길동' : 50, '최창식' : 200 }
temp2 = pd.Series(tempDic)

print(temp + temp2)
```

```python
# loc 조건 활용

temp = { '번호' : ['1번',' 2번', '3번', '4번', '5번'], 
         '이름' : ['홍길동', '전우치', '강감찬', '홍길동', '일지매'],
         '매출' : [100, 200, 250, 300, 150]
       }
tempL = pd.DataFrame(temp)
tempL

# 1. 매출이 100 이상이면서 200 이하인 행만 출력
# tempL.loc[ (tempL['매출'] >= 100) & (tempL['매출'] <= 200) ]

# 2. 매출이 100 이상이면서 200 이하인 행의 이름 컬럼과 매출 컬럼만 출력
# tempL.loc[ (tempL['매출'] >= 100) & (tempL['매출'] <= 200) , ['이름','매출']]

# 3. 매출이 100 이상이거나 200 이하인 행의 이름 컬럼과 매출 컬럼만 출력
tempL.loc[ (tempL['매출'] >= 100) & (tempL['매출'] <= 200), ['이름', '매출'] ]
```

## 행 삭제할 때는 .drop([indexNum])

```python
# 특정 조건에 해당하는 행 삭제. 호날두 행 삭제하기.

# result = tempDel.drop( tempDel[ tempDel['이름'] == '호날두' ].index )
result = tempDel.drop( tempDel[ tempDel.이름 == '호날두' ].index )
result
```

## 해석 : tempDel.drop( tempDel[ tempDel['이름'] == '호날두' ].index ) 
### 1. tempDel['이름'] == '호날두' : 번호 컬럼이 호날두인 애를 찾는다. 이것의 결과값은 번호 컬럼의 모든 행을 조사해서 4번이면 True, 아니면 False 값을 반환한다.
### 2. tempDel[ contents ].index : contents에 맞는 인덱스 값을 리스트 형식( ex, [3] )으로 반환한다. 1번과 합치면 True인 값의 인덱스를 반환하는거지.
### 3. 결국 tempDel.drop( [3] ) 이 되는거임!!

```python
tempDel
```

```python
# 두 개 이상의 행 삭제

result = tempDel.drop( tempDel[ (tempDel.이름 == '홍길동') | (tempDel.지역 == '유벤투스') ].index )
result
```
