# Series 연습

```python
import pandas as pd
s1 = pd.Series(['홍길동', '전우치', '강감찬', '스티브잡스'])
s1
```

```python
# 위 결과에서 인덱스 변경
s1.index = ['1번', '2번', '3번', '4번']
s1
```

```python
# 딕셔너리 형으로 Series 생성하기
sDic = {'일지매' : 120, '전우치' : 100, '홍길동' : 80}
s2 = pd.Series(sDic)
s2
```

```python
# Series 전체 조회
s3 = s2.copy()
print(s3)
print("\n")

# 행 번호로 조회
print(s3[1])
print("\n")

# 인덱스 명으로 조회
print(s3['전우치'])
print("\n")

# 행 번호와 인덱스 명으로 여러 로우 조회
print(s3[['일지매', '전우치', '홍길동']])
print("\n")
print(s3[[0,1,2]])
```

```python
# Series끼리 연산

s4 = s3 + s3
s4
```

# DataFrame 연습

```python
# Dictionary형 변수를 대입하여 DataFrame 만들기

import pandas as pd

dDic = {'번호' : ['1번', '2번', '3번'],
        '생일' : [1975, 1980, 1992],
        '이름' : ['홍길동', '전우치', '강감찬']}
df1 = pd.DataFrame(dDic)
df1
```

```python
# Dictionary형 변수를 대입하여 DataFrame 만들 때, 컬럼 순서 지정

df2 = pd.DataFrame(df1, columns =['이름', '생일', '번호'])
df2
```

```python
# List형을 사용하여 DataFrame 만들기
no = ['1번', '2번', '3번']
name = ['홍길동', '전우치', '강감찬']
birth = [1975, 1982, 1992]

df3 = pd.DataFrame()
df3['번호'] = no
df3['이름'] = name
df3['생일'] = birth
df3
```

```python
# DataFrame 특정 열 기준으로 오름차순 정렬

df4 = df3.copy()
df4.sort_values(['번호'], ascending = True, inplace = True)
df4
```

```python
# loc 조건 활용

temp = { '번호' : ['1번',' 2번', '3번', '4번', '5번'], 
         '이름' : ['홍길동', '전우치', '강감찬', '홍길동', '일지매'],
         '매출' : [100, 200, 250, 300, 150]
       }
df7 = pd.DataFrame(temp)
df7

# 1. 매출이 100 이상이면서 200 이하인 행만 출력
df8 = df7.loc[ (df7.매출 >= 100) & (df7.매출 <= 200)].copy()
df8

# 2. 매출이 100 이상이면서 200 이하인 행의 이름 컬럼과 매출 컬럼만 출력
df9 = df7.loc[ (df7.매출 >= 100) & ( df7.매출 <= 200 ), ['이름', '매출']].copy()
df9

# 3. 매출이 100 이상이거나 200 이하인 행의 이름 컬럼과 매출 컬럼만 출력
df10 = df7.loc[ (df7.매출 >= 100) | (df7.매출 <= 200), ['이름', '매출']].copy()
df10
```

```python
# 변수 넣어서 DataFrame 만들 때 새로운 컬럼 추가

df11 = pd.DataFrame(df10, columns = ['매출', '이름', '컬럼추가'])
df11
```

```python
# 만들어진 DataFrame에서 컬럼 추가
df12 = df11.copy()
df12['컬럼 두번째 추가'] = df12.index
df12
```

```python
# 새로운 행 추가

df13 = df12.copy()
df13.loc[len(df13.index)] = [1, 2, 100, 4]
df13
```

```python
# 2

p2 = pd.read_excel("C:\py_temp\sampleFile\부품구입대장.xlsx", usecols = 'A, C, D, F')

# p2.loc[2][0], ... [2][3] 을 p2.columns[0], ... ,[3] 에 넣기

i = 0
while i < len(p2.columns) :
    p2.rename( columns = {p2.columns[i] : p2.loc[2][i]}, inplace = True ) 
    i += 1
p2.rename(columns = {'신청날짜' : '날짜'}, inplace = True)

p2 = p2.loc[3:]
p2.reset_index(drop=True, inplace = True)
p2
```
