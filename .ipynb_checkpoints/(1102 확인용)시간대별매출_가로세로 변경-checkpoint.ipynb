```python
import pandas as pd
import numpy as np
import os
import sys
import re

file_name = input("전처리할 파일명을 적으시오(ex.시간대별매출_석적점):")
```

```python
df=pd.read_excel('C:\\yehuikim\\2020summer\\프로젝트2\\데이터셋\\feature_data\\'+file_name+'.xls')
# df=pd.read_excel('C:\\yehuikim\\2020summer\\프로젝트2\\데이터셋\\feature_data\\'+file_name+'.xlsx')
df.head()
```

```python
df = df.set_index('판매일자')
df.head()

kk=len(df)
df = df.drop([df.index[kk-1]])
df.tail()

df = df.T
df.tail()

kk=len(df)
df2 = df.drop([df.index[kk-1]])
df2.tail()

df3 = df2.drop([df2.index[0]])
df3.head()

df4 = df3.copy()
col4 = df4.columns.tolist()

for num in range(0,len(col4)) : 
    df4['날짜'+str(num)] = col4[num]
    col_n = df4.columns.tolist()
    tmp = col_n[len(col_n)-1]
    del col_n[len(col_n)-1]
    df4.drop(['날짜'+str(num)],axis='columns')
#     df4.drop([df4.index[len(col4)]])
    col_n.insert(2*num,tmp)
    df4=df4[col_n]
df4.head()

refine_df = pd.DataFrame()
for i in range(1, len(df4.columns), 2):
    temp_df = pd.DataFrame(df4.iloc[:, i])
    temp_df.columns = ['매출액']
    refine_df = pd.concat([refine_df, temp_df])
    print(i, '/', len(df4.columns))
    
date_df = pd.DataFrame()
for i in range(0, len(df4.columns), 2):
    temp_df = pd.DataFrame(df4.iloc[:, i])
    temp_df.columns = ['날짜']
    date_df = pd.concat([date_df, temp_df])
    
final = pd.concat([date_df, refine_df], axis = 1)
```

```python
final.to_excel(file_name+'_전처리완료.xlsx')
```
